```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt
from statistics import mean, stdev, median, variance
import os
from tabulate import tabulate;
import numpy as np
```

```python
path_phasep = './AA/AAS_all_phasenet_p_picks.csv'
path_phases = './AA/AAS_all_phasenet_s_picks.csv'
path_handp = './AA/AAS_all_hand_p_picks.csv'
path_hands = './AA/AAS_all_hand_s_picks.csv'
path_arpickp = './AA/AAS_all_arpick_p_picks.csv'
path_arpicks = './AA/AAS_all_arpick_s_picks.csv'

TPlim = 0.30
Pplim = 0.60
Pslim = 0.40
```

```python
path_project_home = '/Users/utkuunal/Desktop/Projects/ProjectA'
os.chdir('{}'.format(path_project_home))
```

```python
phasep = []
with open(path_phasep) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        phasep.append(row)
print(len(phasep))
print(phasep[:10])
phases = []
with open(path_phases) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        phases.append(row)
print(len(phases))
print(phases[:10])
```

```python
arpickp = []
with open(path_arpickp) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        arpickp.append(row)
print(len(arpickp))
print(arpickp[:10])
arpicks = []
with open(path_arpicks) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        arpicks.append(row)
print(len(arpicks))
print(arpicks[:10])
```

```python
handp = []
with open(path_handp) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        handp.append(row)
print(len(handp))
print(handp[:10])
hands = []
with open(path_hands) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        hands.append(row)
print(len(hands))
print(hands[:10])
```

```python
import collections
phasep_lim = 0
for pp in phasep:
    if float(pp[3]) >= Pplim:
        phasep_lim += 1
phases_lim = 0
for ps in phases:
    if float(ps[3]) >= Pslim:
        phases_lim += 1        
print("{:<68} {:<5} {:<5}".format('Handpicks P, S: ', str(len(handp)) + ',', len(hands)))
print("{:<68} {:<5} {:<5}".format('PhaseNet picks P, S: ', str(phasep_lim) + ',', str(phases_lim)))
phaseeqs = []
i = 0
for pp in phasep:
    if float(pp[3]) >= Pplim:
        for ps in phases:
            if float(ps[3]) >= Pslim:
                if pp[0] == ps[0]:
                    phaseeqs.append(pp[0].split('/')[0])
                    i += 1
print("{:<68} {:<15}".format('PhaseNet P picks matching PhaseNet S picks (by station): ', i))
i = 0
for hs in hands:
    for ps in phases:
        if float(ps[3]) >= Pslim:
            if hs[0] == ps[0]:
                i += 1
print("{:<68} {:<15}".format('Handpicked S picks matching PhaseNet S picks (by station)', i))
print("{:<68} {:<15}".format('Extra Handpicked S picks (Not matching PhaseNet s picks by station)', len(hands)-i))
print("{:<68} {:<15}".format('Extra PhaseNet S picks (Not matching Handpicked s picks by station)', phases_lim-i))
handpickeqs = []
hpstas = []
hpeqs = []
i = 0
for hp in handp:
    if hp[0].split('/')[0] not in hpeqs:
        hpeqs.append(hp[0].split('/')[0])
    if hp[0].split('/')[1] not in hpstas:
        hpstas.append(hp[0].split('/')[1])
    for hs in hands:
        if hp[0] == hs[0]:
            handpickeqs.append(hp[0].split('/')[0])
            i += 1
print(len(handpickeqs), len(hpeqs), len(hpstas))
i = [item for item, count in collections.Counter(handpickeqs).items() if count > 0]
print("{:<68} {:<15}".format('Number of EQs in Handpicked P-S pairs : ', len(i)))
i = [item for item, count in collections.Counter(handpickeqs).items() if count > 2]
print("{:<68} {:<15}".format('Number of EQs in Handpicked P-S pairs with at least 3 stations: ', len(i)))
i = [item for item, count in collections.Counter(phaseeqs).items() if count > 0]
print("{:<68} {:<15}".format('Number of EQs in PhaseNet P-S pairs : ', len(i)))
i = [item for item, count in collections.Counter(phaseeqs).items() if count > 2]
print("{:<68} {:<15}".format('Number of EQs in PhaseNet P-S pairs with at least 3 stations: ', len(i)))
```

```python
peqs = []
for hp in handp:
    if hp[0].split('/')[0] not in peqs:
        peqs.append(hp[0].split('/')[0])
print(len(peqs))
pstations = []
for hp in handp:
    if hp[0].split('/')[1] not in pstations:
        pstations.append(hp[0].split('/')[1])
print(len(pstations))
```

```python
seqs = []
for hs in hands:
    if hs[0].split('/')[0] not in seqs:
        seqs.append(hs[0].split('/')[0])
print(len(seqs))
sstations = []
for hs in hands:
    if hs[0].split('/')[1] not in sstations:
        sstations.append(hs[0].split('/')[1])
print(len(sstations))
```

```python
x_p = []
pp_TP = 0
pp_FP = 0
pp_FN = 0
pp_TN = 0
for pp in phasep:
    for hp in handp:
        if pp[0] == hp[0]:
            a = datetime.strptime(hp[1], '%Y-%m-%dT%H:%M:%S.%f')
            b = datetime.strptime(pp[1], '%Y-%m-%dT%H:%M:%S.%f')
            diff = a - b
            if float(pp[3]) >= Pplim:
                x_p.append(diff.total_seconds())
                if abs(diff.total_seconds()) <= TPlim:
                    pp_TP += 1
                else:
                    pp_FP += 1
            else:
                if abs(diff.total_seconds()) <= TPlim:
                    pp_FN += 1
                else:
                    pp_TN += 1

print('PhaseNet P true positives TP: ', pp_TP)
print('PhaseNet P false positives FP: ', pp_FP)
print('PhaseNet P false negatives: FN ', pp_FN)
print('PhaseNet P true negatives: FN ', pp_TN)
print('Total is: ', pp_TP + pp_FP + pp_FN + pp_TN)

pp_precision = pp_TP / (pp_TP + pp_FP)
pp_recall = pp_TP / (pp_TP + pp_FN)
pp_f1 = 2 * (pp_precision * pp_recall) / (pp_precision + pp_recall)

print(len(x_p))
print(x_p[:100])
```

```python
x_s = []
ps_TP = 0
ps_FP = 0
ps_FN = 0
ps_TN = 0
for ps in phases:
    for hs in hands:
        if ps[0] == hs[0]:
            a = datetime.strptime(hs[1], '%Y-%m-%dT%H:%M:%S.%f')
            b = datetime.strptime(ps[1], '%Y-%m-%dT%H:%M:%S.%f')
            diff = a - b
            if float(ps[3]) >= Pslim:
                x_s.append(diff.total_seconds())
                if abs(diff.total_seconds()) <= TPlim:
                    ps_TP += 1
                else:
                    ps_FP += 1
            else:
                if abs(diff.total_seconds()) <= TPlim:
                    ps_FN += 1
                else:
                    ps_TN += 1                
                
print('PhaseNet S true positives TP: ', ps_TP)
print('PhaseNet S false positives FP: ', ps_FP)
print('PhaseNet S false negatives: FN ', ps_FN)
print('PhaseNet S true negatives: FN ', ps_TN)
print('Total is: ', ps_TP + ps_FP + ps_FN + ps_TN)

ps_precision = ps_TP / (ps_TP + ps_FP)
ps_recall = ps_TP / (ps_TP + ps_FN)
ps_f1 = 2 * (ps_precision * ps_recall) / (ps_precision + ps_recall)

print(len(x_s))
print(x_s[:100])
```

```python
z_p = []
ap_TP = 0
ap_FP = 0
for ap in arpickp:
    for hp in handp:
        if ap[0] == hp[0]:
            a = datetime.strptime(hp[1], '%Y-%m-%dT%H:%M:%S.%f')
            b = datetime.strptime(ap[1], '%Y-%m-%dT%H:%M:%S.%f')
            diff = a - b
            z_p.append(diff.total_seconds())
            if abs(diff.total_seconds()) <= TPlim:
                ap_TP += 1
            else:
                ap_FP += 1
                
print('ARPick P true positives TP: ', ap_TP)
print('ARPick P false positives FP: ', ap_FP)
ap_FN = len(handp) - ap_TP
print('ARPick P false negatives: FN ', ap_FN)

ap_precision = ap_TP / (ap_TP + ap_FP)
ap_recall = ap_TP / (ap_TP + ap_FN)
ap_f1 = 2 * ap_precision * ap_recall / (ap_precision + ap_recall)

print(len(z_p))
print(z_p[:100])
```

```python
z_s = []
as_TP = 0
as_FP = 0
for ars in arpicks:
    for hs in hands:
        if ars[0] == hs[0]:
            a = datetime.strptime(hs[1], '%Y-%m-%dT%H:%M:%S.%f')
            b = datetime.strptime(ars[1], '%Y-%m-%dT%H:%M:%S.%f')
            diff = a - b
            z_s.append(diff.total_seconds())
            if abs(diff.total_seconds()) <= TPlim:
                as_TP += 1
            else:
                as_FP += 1
                
print('ARPick S true positives TP: ', as_TP)
print('ARPick S false positives FP: ', as_FP)
as_FN = len(hands) - as_TP
print('ARPick S false negatives: FN ', as_FN)

as_precision = as_TP / (as_TP + as_FP)
as_recall = as_TP / (as_TP + as_FN)
as_f1 = 2 * as_precision * as_recall / (as_precision + as_recall)

print(len(z_s))
print(z_s[:100])
```

```python
x_p_mean = mean(x_p)
x_p_mae = abs(x_p_mean)
x_p_median = median(x_p)
x_p_stdev = stdev(x_p)
x_p_var = variance(x_p)
x_p_rmse = mean([i ** 2 for i in x_p]) ** 0.5
x_s_mean = mean(x_s)
x_s_mae = abs(x_s_mean)
x_s_median = median(x_s)
x_s_stdev = stdev(x_s)
x_s_var = variance(x_s)
x_s_rmse = mean([i ** 2 for i in x_s]) ** 0.5
```

```python
z_p_mean = mean(z_p)
z_p_mae = abs(z_p_mean)
z_p_median = median(z_p)
z_p_stdev = stdev(z_p)
z_p_var = variance(z_p)
z_p_rmse = mean([i ** 2 for i in z_p]) ** 0.5
z_s_mean = mean(z_s)
z_s_mae = abs(z_s_mean)
z_s_median = median(z_s)
z_s_stdev = stdev(z_s)
z_s_var = variance(z_s)
z_s_rmse = mean([i ** 2 for i in z_s]) ** 0.5
```
